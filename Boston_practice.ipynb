```python
import pandas as pd
import numpy as np
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import statsmodels
import statsmodels.api as sm
import seaborn as sns
boston = sklearn.datasets.load_boston()
boston
```

```python
df= pd.DataFrame(boston['data'])
df
```

```python
df.columns=boston['feature_names']
```

```python
y=boston['target']
x=df
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=5)
```

```python
model= LinearRegression().fit(x_train,y_train)
```

```python
y_hat=model.predict(x_test)
y_hat_train=model.predict(x_train)
```

```python
Eout=((y_test-y_hat)**2).sum()/len(y_test)
Ein=((y_train-y_hat_train)**2).sum()/len(y_train)
print('Eout=',Eout,'\n','Ein=',Ein)
```

以下為透過stats模組

```python
new_x_train=sm.add_constant(x_train)
new_x_test =sm.add_constant(x_test)
new_x_train
```

```python
new_model=sm.OLS(y_train,new_x_train).fit()
```

```python
print(new_model.summary())
```

```python
new_y_hat=new_model.predict(new_x_test)
new_y_hat_train=new_model.predict(new_x_train)
```

```python
new_Eout=((y_test-new_y_hat)**2).sum()/len(y_test)
new_Ein=((y_train-new_y_hat_train)**2).sum()/len(y_train)
```

```python
print('new_Eout=',new_Eout,'\n','new_Ein=',new_Ein)
```
